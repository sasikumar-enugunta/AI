# Objective

To predict whether a patient survived/not survived by urdergoing surgery of breast cancer.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

haberman = pd.read_csv("/home/esk/Downloads/AppliedAI/Module 2: Data Science: Exploratory Data Analysis and Data Visualization/1. Plotting for exploratory data analysis (EDA)/haberman.csv")
```

```python
print(haberman.shape)
```

```python
print(haberman.columns)
```

```python
#Reference from 'https://www.kaggle.com/premvardhan/exploratory-data-analysis-haberman-s-survival'
haberman = haberman.rename(columns = {"30" : "age", "64" : "operation_year", "1" : "axillary_lymph_node", "1.1" : "survival_status"})
print(haberman.columns)
```

```python
haberman = haberman.iloc[1:]
print(haberman)
```

```python
haberman["survival_status"].value_counts()
```

# 2-D Scatter Plot

```python
haberman.plot(kind='scatter', x='age', y='axillary_lymph_node')
plt.title("2-D scatter plot")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(haberman, hue="survival_status", height=4).map(plt.scatter, "age", "axillary_lymph_node").add_legend()
plt.title("3-D scatter plot")
plt.show()
```

#### Observation(s):

    1. Using age and axillary_lymph_node, we can find that axillary_lymph_node between 0-10 survived most.

# Pair-plot

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(haberman, hue="survival_status", vars = ["age", "operation_year", "axillary_lymph_node"], height=3).add_legend()
plt.title("Pair plot")
plt.show()
```

#### Observation(s):
    
    1. We can't classify any information from here as data is not well seperated.

# Histogram, PDF, CDF

```python
sns.FacetGrid(haberman, hue="survival_status", height=5).map(sns.distplot, "age").add_legend()
plt.title("Histogram of age")
plt.show()
```

```python
sns.FacetGrid(haberman, hue="survival_status", height=5).map(sns.distplot, "operation_year").add_legend()
plt.title("Histogram of operation_year")
plt.show()
```

```python
sns.FacetGrid(haberman, hue="survival_status", height=5).map(sns.distplot, "axillary_lymph_node").add_legend()
plt.title("Histogram of axillary_lymph_node")
plt.show()
```

```python
count, bin_edges = np.histogram(haberman['age'], bins=10, density=True)
pdf = count/(sum(count))
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)


count, bin_edges = np.histogram(haberman['age'], bins=20, density = True)

pdf = count/(sum(count))
plt.plot(bin_edges[1:],pdf)

plt.title("PDF & CDF of age")
plt.show()
```

```python
count, bin_edges = np.histogram(haberman['age'], bins=20, density=True)
pdf = count/(sum(count))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)

count, bin_edges = np.histogram(haberman['operation_year'], bins=20, density=True)
pdf = count/(sum(count))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)

count, bin_edges = np.histogram(haberman['axillary_lymph_node'], bins=20, density=True)
pdf = count/(sum(count))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.title("PDF & CDF")
```

# Box plot and Whiskers

```python
sns.boxplot(x = "survival_status", y = "operation_year", hue = "survival_status", data = haberman)
plt.title("Box Plot")
plt.show()
```

# Violin plots

```python
sns.violinplot(x="survival_status", y="age", hue="survival_status", data=haberman, size=8)
plt.title("Violin Plot")
plt.show()
```

# Multivariate probability density, contour plot

```python
#2D Density plot, contors-plot
sns.jointplot(x="survival_status", y="axillary_lymph_node", data=haberman, kind="kde")
plt.title("Contour Plot")
plt.show()
```
