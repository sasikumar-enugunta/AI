<h3> Q5: Find the closest points</h3>

Consider you are given n data points in the form of list of tuples like S=[(x1,y1),(x2,y2),(x3,y3),(x4,y4),(x5,y5),..,(xn,yn)] and a point P=(p,q) <br> your task is to find 5 closest points(based on cosine distance) in S from P

<br>Cosine distance between two points (x,y) and (p,q) is defined as $cos^{-1}(\frac{(x\cdot p+y\cdot q)}{\sqrt(x^2+y^2)\cdot\sqrt(p^2+q^2)})$
<pre>
Ex:

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1)(6,0),(1,-1)]
P= (3,-4)

Output:
(6,-7)
(1,-1)
(6,0)
(-5,-8)
(-1,-1)
</pre>

```python
import math

# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input examples
# you can free to change all these codes/structure


# here S is list of tuples and P is a tuple ot len=2
def closest_points_to_p(S, P):
    # write your code here
    S.sort(key=lambda i: math.sqrt((float(i[0]) - P[0]) ** 2 + (float(i[1]) - P[1]) ** 2))
    return S[:5]

S= [(1,2),(3,4),(-1,1),(6,-7),(0, 6),(-5,-8),(-1,-1),(6,0),(1,-1)]
P= (3,-4)
points = closest_points_to_p(S, P)
print(points) #print the returned values
```

```python
# write your python code here
# you can take the above example as sample input for your program to test
# it should work for any general input try not to hard code for only given input strings


# you can free to change all these codes/structure
def curve_smoothing(string):
    # your code
    list1 = string.split(",")

    idx = prev_idx = prev_val = current_val = 0
    
    while idx < len(list1):
        if list1[idx] != '_' or idx+1 == len(list1):
            value = 0
            if list1[idx] != '_':
                value = int((int(list1[idx]) + prev_val) / (idx - prev_idx + 1))
            elif value == 0:
                value = int(prev_val / (idx - prev_idx + 1))
                
            while prev_idx <= idx:
                list1[prev_idx] = value
                prev_idx += 1
                
            prev_idx -= 1 
            prev_val = value
              
        idx += 1
        
    return list1


S = ["_,_,_,24", "40,_,_,_,60", "80,_,_,_,_", "_,_,30,_,_,_,50,_,_"]
for i in range(len(S)):
    smoothed_values= curve_smoothing(S[i])
    print(smoothed_values)
```
